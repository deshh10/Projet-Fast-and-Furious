<u>Fonction Odeint</u> :

La fonction ‘odeint’ est une fonction du module ‘ scipy.integrate’ de la bibliothèque SciPy en Python. Elle est utilisée pour résoudre des systèmes d'équations différentielles ordinaires (EDO) avec des conditions initiales. Le nom ‘odeint’ signifie "ordinary differential equations integration" (intégration des équations différentielles ordinaires).

Les paramétres de la fonction:

	func : La fonction qui définit le système d'équations différentielles.
	y0 : Les conditions initiales.
	t : Le vecteur des temps auxquels vous souhaitez évaluer la solution.
	args : Les arguments supplémentaires à passer à la fonction func.
	Dfun : La dérivée de func par rapport à y (jacobienne), optionnelle.
	col_deriv : Indique si Dfun renvoie la jacobienne par colonne.
	full_output : Si True, renvoie également un objet détaillé de sortie.
	ml, mu : Si fournis, permettent d'indiquer des points d'arrêt.
	rtol, atol : Les tolérances relatives et absolues pour le contrôle de l'erreur.
	tcrit : Points de temps critiques où le solver doit stopper.
	h0 : Pas initial du solver.
	hmax, hmin : Pas maximal et minimal du solver.
	ixpr : Si 0, aucune impression supplémentaire. Si 1, imprime la ligne de chaque itération.
	mxstep : Nombre maximal de pas.
	mxhnil : Nombre maximal de messages d'avertissement sur la suppression des pas.
	mxordn, mxords : Ordres maximaux pour le solver (mxordn pour la dérivée et mxords pour la solution).
	printmessg : Si True, imprime des messages informatifs.
	tfirst : Si True, l'argument t est le premier argument pour func.

La fonction odeint est utilisée pour résoudre un système d'Équation différentielle de la forme dx/dt=f(x,t) où x est le vecteur des variables dépendantes et t est le temps.

##### Similuation (AVEC FROTTEMENT) + Boost

```python
nitro = 1.3

print("En prévision du passage du ravin souhaitez vous avtivez le Booster NOS ? ")
rep_boost = int(input("1 : Oui \n2 : Non\n"))

if rep_boost == 1: 
    am = am * nitro



# Conditions initiales pour la pente
conditions_initiales_pente = [0, 0]  # position initiale, vitesse initiale
temps_pente = np.linspace(0, 10, 100)  # nb points

# Paramètres de la pente
alpha_pente = np.radians(3.7)  # angle de la pente en rad
mu = 0.1  # Coeff de frottement
rho_air_pente = 1.2  # Densité de l'air (bouge pas)
g = 9.81
# Fonction décrivant le système d'équations différentielles avec frottement
def equation_mouvement_pente_avec_frottement(vitesse_temps, temps):
    g = 9.81
    position, vitesse = vitesse_temps
    a = (m * am + m * g * np.sin(alpha_pente) - mu * m * g * np.cos(alpha_pente) - 0.5 * Cx * S * vitesse**2) / m
    return [vitesse, a]


    

# Résolution de l'équation différentielle avec frottement
solution_pente_avec_frottement = odeint(equation_mouvement_pente_avec_frottement, [0, 0.0], temps_pente)

# Trouver le temps correspondant à une distance de 31 mètres
distance_cible = 31  # en mètres

# Extraire la vitesse à la distance cible
vitesse_31m = solution_pente_avec_frottement[np.abs(solution_pente_avec_frottement[:, 0] - distance_cible).argmin(), 1]

print(f"[AVEC FROTTEMENT] >> Vitesse en bas de la pente ({voiture_choisie}): {vitesse_31m} m/s")

# Graph avec frottement
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(temps_pente, solution_pente_avec_frottement[:, 0], label='Position avec frottement')
plt.axvline(x=temps_pente[np.abs(solution_pente_avec_frottement[:, 0] - distance_cible).argmin()], color='red', linestyle='--', label='31 mètres')
plt.plot(temps_pente, solution_pente_avec_frottement[:, 1], label='Vitesse avec frottement')
plt.xlabel('Temps')
plt.ylabel('Position / Vitesse')
plt.legend()
plt.title(f'Simulation - Pente avec Frottement ({voiture_choisie})')
plt.tight_layout()
plt.show()

```

#### 2 - Le tracé de la vitesse de la voiture au cours du temps<a id="ii2">

##### Paramètres du Looping

```python
rayon_looping = 6.0  # rayon du looping 
g = 9.81  # constante gravitationelle
rho_air = 1.2  # Densité de l'air (bouge pas)
k = 0.5 * Cx * rho_air * S  # Coefficient de traînée
```

##### Conditions Initiales pour la Simulation du Looping


```python
# Conditions initiales pour la simulation du looping
conditions_initiales_looping = [0, vitesse_31m/rayon_looping]  # position initiale, vitesse initiale
temps_simulation_looping = np.linspace(0, 5, 100)  # nb points 
mu = 0.1  # Coefficient de frottement
am, m, S, Cx, Sz = parametres_voitures[voiture_choisie]
```

##### Simulation du Looping (AVEC FROTTEMENT)

```python
# Équation du mouvement avec frottement dans le looping
def equation_mouvement_looping_avec_frottement(variables, temps):
    theta, vitesse_angulaire = variables
    acceleration_angulaire = (am / rayon_looping) - mu * (g / rayon_looping * np.cos(theta) + vitesse_angulaire**2) - g / rayon_looping * np.sin(theta) - k * rayon_looping / m * vitesse_angulaire**2
    return [vitesse_angulaire, acceleration_angulaire]

# Résolution de l'équation avec frottement pour le looping
solution_looping_avec_frottement = odeint(equation_mouvement_looping_avec_frottement, conditions_initiales_looping, temps_simulation_looping)

# Calcul de la vitesse au cours du temps avec frottement pour le looping
vitesse_angulaire_avec_frottement = solution_looping_avec_frottement[:, 1]
vitesse_avec_frottement = rayon_looping * vitesse_angulaire_avec_frottement

# Recherche de la vitesse de sortie du looping et le temps correspondant
index_fin_looping = np.argmax(solution_looping_avec_frottement[:, 0] >= 2 * np.pi)
vitesse_sortie_looping = vitesse_avec_frottement[index_fin_looping]
temps_sortie_looping = temps_simulation_looping[index_fin_looping]

# Graph pour le looping avec frottement
plt.figure(figsize=(12, 6))
plt.plot(temps_simulation_looping, vitesse_avec_frottement, label='Avec frottement')
plt.axvline(x=temps_sortie_looping, color='red', linestyle='--', label='Vitesse de sortie')
plt.xlabel('Temps')
plt.ylabel('Vitesse de la voiture')
plt.legend()
plt.title('Vitesse de la Voiture au cours du Temps - Avec Frottement')
plt.show()

print("[AVEC FROTTEMENT] >> Vitesse de sortie du looping :", vitesse_sortie_looping)
```

##### Simulation du Looping (SANS FROTTEMENT)

```python
# Équation du mouvement sans frottement dans le looping
def equation_mouvement_looping_sans_frottement(vitesse_temps, temps):
    position, vitesse = vitesse_temps
    a = g / rayon_looping - g / rayon_looping * np.sin(position)
    return [vitesse, a]

# Résolution de l'équation sans frottement pour le looping
solution_looping_sans_frottement = odeint(equation_mouvement_looping_sans_frottement, conditions_initiales_looping, temps_simulation_looping)

# Calcul de la vitesse au cours du temps sans frottement pour le looping
vitesse_angulaire_sans_frottement = solution_looping_sans_frottement[:, 1]
vitesse_sans_frottement = rayon_looping * vitesse_angulaire_sans_frottement

# Graph pour le looping sans frottement
plt.figure(figsize=(6, 6))
plt.plot(temps_simulation_looping, vitesse_sans_frottement, label='Sans frottement')
plt.xlabel('Temps')
plt.ylabel('Vitesse de la voiture')
plt.legend()
plt.title('Vitesse de la Voiture au cours du Temps - Sans Frottement')
plt.show()
```

#### 3 - Le calcul de la vitesse minimale d'entrée dans le looping pour faire le tour sans tomber<a id="ii3">

```python
# Calcul de la vitesse minimale d'entrée avec frottement
vitesse_min_entree_avec_frottement = np.sqrt(rayon_looping * (5 * g - 2 * am * np.pi))

# Affichage de la vitesse minimale d'entrée avec frottement
print(f"[AVEC FROTTEMENT] >> Vitesse minimale d'entrée dans le looping pour faire un tour sans tomber : {vitesse_min_entree_avec_frottement} m/s")
```

```python
# Calcul de la vitesse minimale d'entrée dans le looping sans frottement
vitesse_minimale_entree_looping_sans_frottement = np.sqrt(rayon_looping * (5 * g - 2 * am * np.pi))
print(f"[SANS FROTTEMENT] >> Vitesse minimale d'entrée dans le looping pour faire un tour sans tomber : {vitesse_minimale_entree_looping_sans_frottement} m/s")
```

#### 4 - Le calcul de la vitesse de sortie du looping <a id="ii4">

```python
# Conditions initiales pour la sortie du looping avec frottement
conditions_initiales_sortie_looping_avec_frottement = [np.pi, 0]  # position initiale (en bas du looping), vitesse initiale

# Résolution de l'équation avec frottement pour la sortie du looping
solution_sortie_looping_avec_frottement = odeint(equation_mouvement_looping_avec_frottement, conditions_initiales_sortie_looping_avec_frottement, temps_simulation_looping)

# Vitesse de sortie du looping avec frottement
vitesse_sortie_looping = vitesse_avec_frottement[index_fin_looping]
print("[AVEC FROTTEMENT] >> Vitesse de sortie du looping :", vitesse_sortie_looping, "m/s","\n")

# Vitesse de sortie du looping sans frottement
vitesse_sortie_looping_sans_frottement = np.sqrt(rayon_looping * (5 * g + 2 * am * np.pi))
print(f"[SANS FROTTEMENT] >> Vitesse de sortie du looping : {vitesse_sortie_looping_sans_frottement} m/s")
```

```python
# Paramètres du ravin avec frottement
vitesse_initiale_ravin = vitesse_sortie_looping_sans_frottement  # m/s
xmin_ravin = 9.0  # m
h0 = 1
rho = 1.2
Cz = 0.3

# Calcul de la vitesse initiale nécessaire pour le ravin sans frottement
V0 = np.sqrt((xmin_ravin**2) / (2 * h0 * g))

# Conditions initiales pour la simulation du ravin avec frottement
conditions_initiales_ravin = [0, vitesse_sortie_looping_sans_frottement, h0, 0]  # position x, vitesse x, position y, vitesse y
temps_simulation_ravin = np.linspace(0, 5, 100)  # points temporels pour la simulation

# Coefficients pour le ravin avec frottement
k = 0.5 * Cx * rho * S

# Équations du mouvement avec frottement pour le ravin
def equation_mouvement_ravin_avec_frottement(variables, temps):
    px, vx, py, vy = variables
    
    # Équations différentielles pour le ravin avec frottement
    v = np.sqrt(vx**2 + vy**2)
    dvxdt = (v * (-k * vx)) / m
    dvydt = (v * (-k * vy)) / m - g

    return [vx, dvxdt, vy, dvydt]

# Résolution de l'équation avec frottement pour le ravin
solution_ravin_avec_frottement = odeint(equation_mouvement_ravin_avec_frottement, conditions_initiales_ravin, temps_simulation_ravin)

# Graph pour le ravin avec frottement
plt.figure(figsize=(12, 6))
plt.plot(solution_ravin_avec_frottement[:, 0], solution_ravin_avec_frottement[:, 2], label='Trajectoire dans le ravin avec frottement')
plt.axhline(y=0, color='black', linestyle='--', label='Surface du sol')
plt.axvline(x=xmin_ravin, color='red', linestyle='--', label='Extrémité du ravin')
plt.xlabel('Position x')
plt.ylabel('Position y')
plt.legend()
plt.title('Trajectoire de la Voiture dans le Ravin avec Frottement')
plt.xlim(0, 10)  # Zoom sur la plage de 0 à 9 mètres
plage_x = (-0.9, 1.2)
plt.ylim(plage_x[0], plage_x[1])

plt.show()
```

##### Sans Frottement

```python
# Équations du mouvement sans frottement pour le ravin
def equation_mouvement_ravin_sans_frottement(variables, temps):
    vx, ax, vy, ay = variables
    dvxdt = 0  # Pas de frottement sur l'axe x
    dvydt = -g  # Accélération due à la gravité sur l'axe y
    return [ax, dvxdt, ay, dvydt]

# Résolution de l'équation sans frottement pour le ravin
solution_ravin_sans_frottement = odeint(equation_mouvement_ravin_sans_frottement, conditions_initiales_ravin, temps_simulation_ravin)

# Graph pour le ravin sans frottement
plt.figure(figsize=(12, 6))
plt.plot(solution_ravin_sans_frottement[:, 0], solution_ravin_sans_frottement[:, 2], label='Trajectoire dans le ravin sans frottement')
plt.axhline(y=0, color='black', linestyle='--', label='Surface du sol')
plt.axvline(x=xmin_ravin, color='red', linestyle='--', label='Extrémité du ravin')
plt.xlabel('Position x')
plt.ylabel('Position y')
plt.legend()
plt.title('Trajectoire de la Voiture dans le Ravin sans Frottement')
plt.xlim(0, 10)  # Zoom sur la plage de 0 à 9 mètres

# Utilisez la plage_x pour définir la limite y
plage_x = (-0.9, 1.2)
plt.ylim(plage_x[0], plage_x[1])

plt.show()
```
